```python
#pulling information from one of my favorite youtube channels "GoodBadFlicks" using 
#https://developers.google.com/youtube/v3/docs/search/list 
#to request and sort the information
import requests
import pandas as pd
import time
```

```python
def get_stats(video_id):
    
    url2 = 'https://www.googleapis.com/youtube/v3/videos?id='+video_id+'&part=statistics&key='+API_KEY
    call2 = requests.get(url2).json()

    viewCount = call2['items'][0]['statistics']['viewCount']
    likeCount = call2['items'][0]['statistics']['likeCount']
    dislikeCount = call2['items'][0]['statistics']['dislikeCount']
    commentCount = call2['items'][0]['statistics']['commentCount']
    
    return viewCount, likeCount, dislikeCount, commentCount
```

```python
def get_videos(df):
    
    pageToken = ""
    url = 'https://www.googleapis.com/youtube/v3/search?key='+API_KEY+'&channelId='+CHANNEL_ID+'&part=snippet,id&order=viewCount&maxResults=10000'+pageToken
    call  = requests.get(url).json()
    
    time.sleep(1)

    for video in call['items']:
        if video['id']['kind'] == 'youtube#video':
            video_id = video['id']['videoId']
            title = video['snippet']['title']
            date_pub = video['snippet']['publishedAt']
            date_pub = str(date_pub).split('T')[0]

            viewCount,likeCount,dislikeCount,commentCount = get_stats(video_id)

            df = df.append({'video_id': video_id, 'title': title, 'date_pub':date_pub, 
                            'view_count':viewCount,'like_count':likeCount,
                            'dislike_count':dislikeCount, 'comment_count': commentCount}, ignore_index=True)
        

    return df
```

```python
df = pd.DataFrame(columns=['video_id','title','date_pub','view_count','like_count','dislike_count','comment_count'])
df = get_videos(df)
```
